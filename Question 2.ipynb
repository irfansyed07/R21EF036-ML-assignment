```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/HP/OneDrive/Desktop/prices_clean.csv")
```

```python
make_prices = df.groupby('make')['price'].mean().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='make', y='price', data=make_prices)
plt.xticks(rotation=90)
plt.xlabel('Car Make')
plt.ylabel('Average Price')
plt.title('Average Prices by Car Make in 2023')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Distribution of Car Prices in 2023')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='make', y='price', data=df)
plt.xticks(rotation=90)
plt.xlabel('Car Make')
plt.ylabel('Price')
plt.title('Car Prices by Make in 2023')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='model', y='price', data=df, hue='make')
plt.xticks(rotation=90)
plt.xlabel('Car Model')
plt.ylabel('Price')
plt.title('Car Prices by Model in 2023')
plt.legend(title='Car Make')
plt.show()
```

```python
honda_prices = df[df['make'] == 'Honda']
bmw_prices = df[df['make'] == 'BMW']

plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='price', data=honda_prices, label='Honda')
sns.lineplot(x='year', y='price', data=bmw_prices, label='BMW')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Price Trends Over Time (2023)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(honda_prices['price'], label='Honda', shade=True)
plt.xlabel('Price')
plt.ylabel('Density')
plt.title('KDE Plot of Honda Car Prices in 2023')
plt.legend()
plt.show()
```

```python
make_counts = df['make'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(make_counts, labels=make_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Market Share by Car Make in 2023')
plt.show()
```

```python
top_10_expensive = df.nlargest(10, 'price')

plt.figure(figsize=(12, 6))
sns.barplot(x='model', y='price', data=top_10_expensive)
plt.xticks(rotation=90)
plt.xlabel('Car Model')
plt.ylabel('Price')
plt.title('Top 10 Most Expensive Car Models in 2023')
plt.show()
```
